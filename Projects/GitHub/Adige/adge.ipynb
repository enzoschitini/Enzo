```python
# Categorias

categ_num = []
catg = []
with open(file='.csv', mode='r', encoding='utf8') as file:
    line = file.readline()
    line = file.readline()
    while line:
        line_x = line.split(sep=',')
        line_cat = line.split(sep=',')
        line_cat = line_cat[0]
        catg.append(line_cat)
        line_add = line_x[0] + '-' + line_x[1]
        line_add = line_add.strip()
        categ_num.append(line_add)
        line = file.readline()

# Funções #########################################
###################################################

# Limpar uma lista - tirar valores repetidos
def limp(lista_limp : list) -> list:
    lista_limp = list(set(lista_limp))
    return lista_limp

# Limpar arquivo csv apagar tudo
def limp_csv(arquivo_csv : str):
   with open(file=arquivo_csv, mode='w') as file:
      file.write("")


# Extrair da celula um valor em uma posição
def extcel(lista_divd : list, elemento : int, simb : str, tipo : str) -> list:
   nova_lista = []
   for x in lista_divd:
      x = x.split(sep=simb)
      x = x[elemento]
      x = x.strip()
      if tipo == 'int':
         x = int(x)
         nova_lista.append(x)
      elif tipo == 'float':
         x = float(x)
         nova_lista.append(x)
      elif tipo == 'str':
         x = str(x)
         nova_lista.append(x)
   return nova_lista

# Media total
def media_de_lista(lista_media : list, round_num : int) -> float:
   from functools import reduce
   media = reduce(lambda x, y: x + y, lista_media) # Total
   media = media / len(lista_media) # Dividindo pela quantidade
   media = round(media, round_num)
   return media

valores = extcel(categ_num, 1, '-', 'float') 
media = media_de_lista(valores, 2)
print(media)


##############################################################
##############################################################

print(categ_num)
catg = limp(catg)
print(catg)
listagem = []
for cat in catg:
    for i in categ_num:
     divd = i.split(sep='-')
     if divd[0] == cat:
        listagem.append((divd[1]).strip())

print(listagem)

def second():
   import time
   from time import sleep

   start_time = time.time()
   time_x = round((time.time() - start_time), 2)
   print("--- %s seconds ---" % time_x)
```
