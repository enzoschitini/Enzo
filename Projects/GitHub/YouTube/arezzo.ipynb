```python
# Categorias e Visualiza√ß√µes

lista = []

with open(file='./youtube.csv', mode='r') as file:
    linha = file.readline()
    linha = file.readline()
    while linha:
        linha_x = linha.split(sep=',')
        linha_add = linha_x[4]
        lista.append(linha_add)
        linha = file.readline()

filtered_data = [item for item in lista if 'nan' not in item]
categorias = filtered_data
print(categorias)

todos = []

with open(file='./youtube.csv', mode='r') as file:
    linha = file.readline()
    linha = file.readline()
    while linha:
        linha_x = linha.split(sep=',')
        category = linha_x[4]
        valor = linha_x[3]
        valor_sep = list(valor)
        for x in valor_sep:
            if x != '+':
                f = category + ', ' + valor
                todos.append(f)
                linha = file.readline()
        linha = file.readline()

todos = [item for item in todos if '+' not in item]
print(todos)
```

```python
def categoria(lista : list, catg : list) -> list:
   total = []
   for x in catg:
      soma = sum(int(item.split(', ')[1]) for item in lista if item.split(', ')[0] == x)
      total.append(soma)
   return total

def limpstring(string : str, elem : str) -> str:
   lista = list(string)
   cancel = 0
   for i in lista:
      if i == elem:
         cancel = 1
   return cancel
```
